```python
import pandas as pd 
import numpy as np 
df1=pd.read_csv(r'C:\Users\vaish\Documents\Vaishnavi_datascience\Movie_Recommendation\movie2\tmdb_5000_credits.csv')
df2=pd.read_csv(r'C:\Users\vaish\Documents\Vaishnavi_datascience\Movie_Recommendation\movie2\tmdb_5000_movies.csv')
```

```python
df1.columns = ['id','tittle','cast','crew']
df2= df2.merge(df1,on='id') 
#df2 shows the movie_data
```

```python
df2.head(5)
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df2.shape
     
```

```python

# selecting the relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast']
print(selected_features)
```

```python
combined_features = df2['genres']+' '+df2['keywords']+' '+df2['tagline']+' '+df2['cast']+' '
     
```

```python
print(combined_features)
     
```

```python
# converting the text data to feature vectors i.e td-idf matrix

vectorizer = TfidfVectorizer()
```

```python
# Remove rows with NaN values
combined_features = pd.Series(combined_features).dropna().tolist()

# Now you can fit_transform
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
     
```

```python
print(similarity)
```

```python
print(similarity.shape)     
```

```python
# getting the movie name from the user

movie_name = input(' Enter your favourite movie name : ')
```

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python

movie_name = input('Enter your favourite movie name: ')

list_of_all_titles = df2['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

if find_close_match:
    close_match = find_close_match[0]
    index_of_the_movie = df2[df2.title == close_match].index[0]  # Get the index of the close match

    if index_of_the_movie < similarity.shape[0]:
        
        similarity_score = list(enumerate(similarity[index_of_the_movie]))
        
        # Create a list of recommended movies with their similarity scores
        recommended_movies = []

        # Check for movies that contain the input title and add them to the recommended list
        for movie in similarity_score:
            index = movie[0]
            title_from_index = df2.iloc[index]['title']  # Get movie title
            score = movie[1]

            # Add the movie to the recommended list
            recommended_movies.append((title_from_index, score))

        prioritized_movies = sorted(recommended_movies, key=lambda x: (movie_name.lower() in x[0].lower(), x[1]), reverse=True)

    
        print('Movies suggested for you : \n')
        for i, (title, score) in enumerate(prioritized_movies):
            if i < 30:
                print(f"{i + 1}. {title}")

    else:
        print(f"No close matches found for {movie_name}")
else:
    print(f"No close matches found for {movie_name}")
```
